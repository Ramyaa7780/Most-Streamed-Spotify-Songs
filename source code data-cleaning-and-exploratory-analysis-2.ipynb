#### Dealing with the data and NaNs

A lot of features like spotify, tiktok, apple music, youtube metrics have object type of data where they clearly should have numerical values. This must be fixed because it might cause problems later on with the analysis.

```python
stream_service_metric_names = [
    'spotify_streams','spotify_playlist_count','spotify_playlist_reach',
    'youtube_views','youtube_likes','youtube_playlist_reach',
    'tiktok_posts', 'tiktok_likes', 'tiktok_views',
    'airplay_spins','siriusxm_spins','deezer_playlist_reach','pandora_streams', 'soundcloud_streams',
    'shazam_counts'
]
```

```python
spotify_data[stream_service_metric_names].apply(lambda x: x.str.replace(',',''))
```

```python
spotify_data[stream_service_metric_names] = spotify_data[stream_service_metric_names].apply(lambda x: x.str.replace(',',''))
```

```python
spotify_data[stream_service_metric_names].dropna().astype(int)
```

The easiest way to deal with the NaNs would be to simply remove the rows with missing data. However this would cause a massive loss on available data for other features. I am left with almost mere 13% of the whole data. I don't like how I loose tons of information when I drop the null rows. In that case I would rather fill in the NaN values. For that purpose I am going to youse the `pd.interpolate()` method to replace the missing values with the closest one to them. After that I can proceed with analyzing. For that I will also have to cast these columns as `Int64`

```python
spotify_data[stream_service_metric_names].apply(lambda x: pd.to_numeric(x)).astype('Int64')
```

```python
spotify_data[stream_service_metric_names] = spotify_data[stream_service_metric_names].apply(lambda x: pd.to_numeric(x,errors='coerce')).astype('Int64')
```

```python
spotify_data[stream_service_metric_names] = spotify_data[stream_service_metric_names].interpolate(method='ffill')
```

### Problem 6. Playlists (2 points)
Is there any connection (correlation) between users adding a song to playlists in one service, or another? Only Spotify, Apple, Deezer, and Amazon offer the ability to add a song to a playlist. Find a way to plot all these relationships at the same time, and analyze them. Experiment with different types of correlations.

```python
stream_providers = ['spotify_playlist_count','apple_music_playlist_count','deezer_playlist_count','amazon_playlist_count']
```

```python
playlist_count_data = spotify_data[stream_providers]
```

```python
playlist_count_data = playlist_count_data.interpolate(method='ffill').astype(int)
```

```python
playlist_count_data
```

In the current state the data is set in different ranges. Spotify playlists vary between a 4-5 thousand to a 30 thousand.
Meanwhile apple,deezer and amazon are set to a lower range between 1 and a 100-200. This would make the data unfit to work with and I will have to provide some sort normalization in order to work on the same scale. For this purpose I am going to use the `MinMaxScaler` in order to scale the data down to a range `[0 ; 1]`

```python
scaler = MinMaxScaler()
```

```python
playlist_count_data = pd.DataFrame(scaler.fit_transform(playlist_count_data),columns=stream_providers)
```

```python
playlist_count_data
```

To choose which type of correlation test I want to use first I am going to look for outliers. In this case I am going to use a boxplot.

```python
playlist_count_data.describe().T
```

After having a look at the described data I have noticed a big difference in the features' means and their max value. This indicates outliers. In that case I will want to use box-plots to have visual representation of the outliers.

```python
plt.figure(figsize=(10,10))

sns.catplot(data = playlist_count_data, kind='box')

plt.xticks(rotation=90)

plt.show()
```

After visualizing the distribution of the features I clearly see that we have an immense amount of outliers for each vector.
This will cause big deviation and influence in correlation methods like 'spearman' and 'pearson' as shown down below.

```python
playlist_count_data.corr() # if empty pearson
```

```python
playlist_count_data.corr('spearman')
```

```python
playlist_count_data.corr('kendall')
```

```python
pearson = playlist_count_data.corr() 
spearman = playlist_count_data.corr('spearman')
kendall = playlist_count_data.corr('kendall')
```

```python
ig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.heatmap(pearson, annot=True, cmap='crest', ax=axes[0])
axes[0].set_title('Pearson Correlation')

sns.heatmap(spearman, annot=True, cmap='crest', ax=axes[1])
axes[1].set_title('Spearman Correlation')

sns.heatmap(kendall, annot=True, cmap='crest', ax=axes[2])
axes[2].set_title('Kendall Correlation')

plt.tight_layout()
plt.show()
```

From all three methods the most robust against outliers is the `Kendal` method for calculating correlations. Therefore I am going to assume its result are the most valid from all three:

Since none of the correlation indecies reaches beyond 0.5 I would't assume that there is any strong correlation between adding a song to a playlist of a different streaming service
